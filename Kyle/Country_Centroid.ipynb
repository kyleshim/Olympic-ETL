```python
import pandas as pd
```

```python
country_code_file = "Resources/wikipedia-iso-country-codes.csv"
country_code_df = pd.read_csv(country_code_file)
```

```python
url = 'https://developers.google.com/public-data/docs/canonical/countries_csv'
```

```python
tables = pd.read_html(url)
tables
```

```python
countries_df = tables[0]
countries_df
```

```python
country_code_df = country_code_df.rename(columns={"English short name lower case":"Country"})
country_code_df
```

```python
countries_df = countries_df.rename(columns={"name":"Country", "country":"Alpha-2 code"})
countries_df = countries_df.sort_values(by="Country")
countries_df
```

```python
centroid_2_df = pd.merge(countries_df, country_code_df, on="Alpha-2 code", how="outer")
centroid_2_df = centroid_2_df.dropna()
centroid_2_df
```

```python
centroid_2b_df = pd.merge(countries_df, 
                          country_code_df, 
                          left_on=["Alpha-2 code","Country"], 
                          right_on=["Alpha-2 code","Country"], 
                          how="outer")
centroid_2b_df = centroid_2b_df.sort_values(by='Alpha-2 code')
centroid_2b_df = centroid_2b_df.drop_duplicates(subset=['Alpha-2 code'])
centroid_2b_df
```

```python
country_x_df = centroid_2_df.dropna(subset=["Country_x","Country_y"])
country_x_df = country_x_df[['Alpha-2 code','latitude','longitude','Country_x','Alpha-3 code','Numeric code','ISO 3166-2']]
country_x_df = country_x_df.rename(columns={'Country_x':'Country'})
country_x_df
```

```python
country_y_df = centroid_2_df.dropna(subset=["Country_y"])
country_y_df = country_y_df[['Alpha-2 code','latitude','longitude','Country_y','Alpha-3 code','Numeric code','ISO 3166-2']]
country_y_df = country_y_df.rename(columns={'Country_y':'Country'})
country_y_df
```

```python
centroid_name_df = pd.merge(countries_df, country_code_df, on="Country", how="outer")
centroid_name_df = centroid_name_df.sort_values(by='Country')
centroid_name_df = centroid_name_df.drop_duplicates(subset=['Alpha-3 code'])
centroid_name_df = centroid_name_df.sort_values(by='Alpha-2 code_y')
centroid_name_df
```

```python
centroid_2_df.loc[centroid_2_df['Country_x'].isnull(),'Country_x_is_NaN'] = 'Yes'
centroid_2_df.loc[centroid_2_df['Country_y'].isnull(),'Country_y_is_NaN'] = 'Yes'
centroid_2_df
```

```python
country_x_nan = centroid_2_df.set_index('Country_x_is_NaN')
country_x_nan = country_x_nan.loc['Yes']
country_x_nan
```

```python
country_y_nan = centroid_2_df.set_index('Country_y_is_NaN')
country_y_nan = country_y_nan.loc['Yes']
country_y_nan
```

```python
centroid_name_df.loc[centroid_name_df['Alpha-2 code_x'].isnull(),'Alpha-2 code_x'] = 'Yes'
centroid_name_df.loc[centroid_name_df['Alpha-2 code_y'].isnull(),'Alpha-2 code_y'] = 'Yes'
centroid_name_df
```

```python
alpha_x_nan = centroid_name_df.set_index('Alpha-2 code_x')
alpha_x_nan = alpha_x_nan.loc['Yes']
alpha_x_nan
```

```python
alpha_y_nan = centroid_name_df.set_index('Alpha-2 code_y')
alpha_y_nan = alpha_y_nan.loc['Yes']
alpha_y_nan
```

```python
centroid_country_df = centroid_2_df[['Alpha-2 code','latitude','longitude','Country_x','Country_y','Alpha-3 code','Numeric code','ISO 3166-2']]
```

```python
centroid_country_df=pd.merge(country_x_df, country_y_df, how='outer', on='Country')
centroid_country_df
```
