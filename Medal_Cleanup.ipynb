```python
import pandas as pd

medal_df = pd.read_csv("Data/medals.csv")
country_code_df = pd.read_csv("Data/wikipedia-iso-country-codes.csv")
```

```python
medal_df
```

```python
medal_all = medal_df.groupby(by = ["NOC", "Year"]).count()
medal_all = medal_all.iloc[:, [0]]
medal_all = medal_all.rename(columns={"City": "All"})
medal_all.head(5)
```

```python
medal_gold = medal_df.loc[medal_df["Medal"] == "Gold"]
medal_gold = medal_gold.groupby(by = ["NOC", "Year"]).count()
medal_gold = medal_gold.iloc[:, [0]]
medal_gold = medal_gold.rename(columns={"City": "Gold"})
medal_gold.head(5)
```

```python
medal_silver = medal_df.loc[medal_df["Medal"] == "Silver"]
medal_silver = medal_silver.groupby(by = ["NOC", "Year"]).count()
medal_silver = medal_silver.iloc[:, [0]]
medal_silver = medal_silver.rename(columns={"City": "Silver"})
medal_silver.head(5)
```

```python
medal_bronze = medal_df.loc[medal_df["Medal"] == "Bronze"]
medal_bronze = medal_bronze.groupby(by = ["NOC", "Year"]).count()
medal_bronze = medal_bronze.iloc[:, [0]]
medal_bronze = medal_bronze.rename(columns={"City": "Bronze"})
medal_bronze.head(5)
```

```python
country_code_df
```

```python
country_code_df = country_code_df.iloc[:, [0,1,2]]
country_code_df
```

```python
modified_medal_df = pd.merge(medal_all, medal_gold, on = ["NOC", "Year"], how = "outer")
modified_medal_df = pd.merge(modified_medal_df, medal_silver, on = ["NOC", "Year"], how = "outer")
modified_medal_df = pd.merge(modified_medal_df, medal_bronze, on = ["NOC", "Year"], how = "outer")
modified_medal_df = modified_medal_df.reset_index()
modified_medal_df
```

```python
modified_medal_df = modified_medal_df.rename(columns = {"NOC": "Alpha-3 code"})
cleaned_medal_df = pd.merge(modified_medal_df, country_code_df, on = ["Alpha-3 code"], how = "left")
cleaned_medal_df
```

```python
cleaned_medal_df.to_csv("Data/cleaned_medal_dataframe.csv")
```
